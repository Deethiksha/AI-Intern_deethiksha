```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("STUDENTS_RAW_DATA (1).csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
placeholders = ["NA", "N/A", "na", "-", "--", "unknown", "Unknown", ""]
```

```python
df.replace(placeholders, np.nan, inplace=True)
```

```python
text_cols = df.select_dtypes(include="object").columns
for col in text_cols:
    df[col] = df[col].str.strip().str.lower()
```

```python
df["admission_date"] = pd.to_datetime(df["admission_date"], errors="coerce")
```

```python
numeric_cols = ["age", "study_hours_per_week","attendance_rate", "prior_gpa_10pt","test_score", "fee_paid_inr"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.loc[df["prior_gpa_10pt"] > 10, "prior_gpa_10pt"] = np.nan
df.loc[df["attendance_rate"] > 100, "attendance_rate"] = np.nan
```

```python
df["study_hours_per_week"] = df["study_hours_per_week"].clip(0, 80)
```

```python
df["attendance_rate"] = df["attendance_rate"].fillna(df["attendance_rate"].median())
```

```python
df_model = pd.get_dummies(df, drop_first=True)
```

```python
df.to_csv("STUDENTS_CLEANED_DATA.csv", index=False)
df_model.to_csv("STUDENTS_MODEL_DATA.csv", index=False)
```
